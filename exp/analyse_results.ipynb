```python
import pickle
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
root = './results/'
experiment = 'IMDBBINARY-1611859425.2614622'
fold = 0
path = os.path.join(root, experiment, 'fold-{}'.format(fold), 'curves.pkl')
```

```python
with open(path, 'rb') as handle:
    curves = pickle.load(handle)
```

```python
train_curve = [item for item in curves['train_loss']]
plt.figure(dpi=300)
plt.plot(train_curve, lw=0.5)
plt.show()
plt.close()
```
